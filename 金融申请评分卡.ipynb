```python
for i in allFeatures:
    print('变量:{}的不同水平值有{}个'.format(i ,len(data[i].unique())))
```

```python
a={}
for i in allFeatures:
    a[i]=len(data[i].unique())
```

```python
a
```

```python
#先把变量分成连续变量与分类变量，分类依据是水平种类在10种以上，归类为连续变量。低于10种一下是分类变量
categorical_var=[]
continuous_var=[]
for key ,values in a.items():
    if values>18:
        continuous_var.append(key)
    else:
        categorical_var.append(key)
```
